```python
from venn import venn
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#UCEC One TMT experiment
#load standardized dataframe
df_maxquant_T=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/df_maxquant_T_Refseq_venn.pkl')
df_maxquant_not_filtered=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/df_maxquant_T_Refseq_not_filtered_venn.pkl')
CDAP=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/CDAP_venn.pkl')
CDAP_not_filtered=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/CDAP_not_filtered_venn.pkl')
PNNL=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/PNNL_venn.pkl')
PNNL_not_filtered=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/PNNL_not_filtered_venn.pkl')
Michi_2=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/Michi_venn.pkl')
Michi_not_filtered=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/Michi_not_filtered_venn.pkl')
CDAP=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/CDAP_w_rt.pkl')
PNNL=pd.read_pickle('../../data/AutoRT_data/UCEC/standardized_df/PNNL_w_rt.pkl')
```

```python
#PSMs from each pipeline
#exp+scan+phosphopep
maxquant_zipped=list(zip(df_maxquant_T.index.tolist(),df_maxquant_T['scan'].tolist(),df_maxquant_T['phosphopep'].tolist()))
CDAP_zipped=list(zip(CDAP.exp.tolist(),CDAP.scan.tolist(),CDAP.phosphopep.tolist()))
Michi_zipped_2=list(zip(Michi_2.index.tolist(),Michi_2.scan.tolist(),Michi_2.phosphopep.tolist()))
PNNL_zipped = list(zip(PNNL.exp.tolist(),PNNL.scan.tolist(),PNNL.phosphopep.tolist()))
```

```python
PSMs = {
    "MaxQuant": set(maxquant_zipped),
    "CDAP": set(CDAP_zipped),
    "FragPipe": set(Michi_zipped_2),
    "MS-GF+/Ascore": set(PNNL_zipped)
}
plt.figure(1, figsize=(12,8))
venn(PSMs,cmap=['dodgerblue','steelblue','royalblue','cornflowerblue'], fontsize=12)
# Show graphic
plt.show()
```

```python
PSMs = {
    "MaxQuant": set(maxquant_zipped),
    "FragPipe": set(Michi_zipped_2),
    "MS-GF+/Ascore": set(PNNL_zipped),
    "CDAP": set(CDAP_zipped)
}
plt.figure(1, figsize=(12,8))
venn(PSMs, fmt="{percentage:.1f}%",cmap=['dodgerblue','steelblue','royalblue','cornflowerblue'], fontsize=12)
plt.title('PSM filtered',position=(0.5,0.85))
# Show graphic
plt.show()
```

```python
maxquant_zipped=list(zip(df_maxquant_not_filtered.index.tolist(),df_maxquant_not_filtered['scan'].tolist(),df_maxquant_not_filtered['sequence'].tolist()))
CDAP_zipped=list(zip(CDAP_not_filtered.exp.tolist(),CDAP_not_filtered.scan.tolist(),CDAP_not_filtered.sequence.tolist()))
Michi_zipped_2=list(zip(Michi_not_filtered.index.tolist(),Michi_not_filtered.scan.tolist(),Michi_not_filtered.sequence.tolist()))
PNNL_zipped = list(zip(PNNL_not_filtered.exp.tolist(),PNNL_not_filtered.scan.tolist(),PNNL_not_filtered.sequence.tolist()))
import matplotlib.pyplot as plt
PSMs = {
    "MaxQuant": set(maxquant_zipped),
    "FragPipe": set(Michi_zipped_2),
    "MS-GF+/Ascore": set(PNNL_zipped),
    "CDAP": set(CDAP_zipped)
}
plt.figure(1, figsize=(12,8))
venn(PSMs, fmt="{percentage:.1f}%",cmap=['dodgerblue','steelblue','royalblue','cornflowerblue'], fontsize=12)
plt.title('PSM no filtered',position=(0.5,0.85))
# Show graphic
plt.show()
```

```python
PSMs = {
    "MaxQuant": set(df_maxquant_T.phosphopep),
    "FragPipe": set(Michi_2.phosphopep),
    "MS-GF+/Ascore": set(PNNL.phosphopep),
    "CDAP": set(CDAP.phosphopep)
}
plt.figure(1, figsize=(12,8))
venn(PSMs, cmap=['dodgerblue','steelblue','royalblue','cornflowerblue'], fontsize=12, legend_loc=(0.8,0.75))
plt.title('phophopeptide filtered',position=(0.5,0.85))
# Show graphic
plt.show()
```
